```python
df = pd.read_csv("./mentions_musk.csv")
```

```python
tweets_list = []
for index in range(len(df.index)):
    txt = df.iloc[index,2]
    tweets_list.append([txt])
```

```python
import re
def text_to_word_list(tweet):
    
    txt = tweet[0]
    
    txt = re.sub('\n', ' ', txt)
    
    txt = re.sub(r'http\S+', '', txt)#removes links
    #txt = re.sub("#([a-zA-Z0-9_]{1,50})", '', txt)#remove hashtags
    
    whitelist = set('abcdefghijklmnopqrstuvwxyz ABCDEFGHIJKLMNOPQRSTUVWXYZ')
    cleaned_tweet = ''.join(filter(whitelist.__contains__, txt))
    
    text = cleaned_tweet.split(" ")
    return text
```

```python
corpus = []
for tweet in tweets_list:
    corpus.append(text_to_word_list(tweet))
```

```python
#df.loc[0,['text']]
tweets_list
```

```python
from gensim.models import Word2Vec
```

```python
model_w2v = Word2Vec(corpus, min_count=1,vector_size = 100,workers=3, window =3, sg = 1)
```

```python
model_w2v.wv["elon"]
```

```python
model_w2v.wv["musk"]
```

```python
model_w2v.wv.similarity('muskrat', 'musk')
```

```python
from sklearn.cluster import KMeans
```

```python
NUM_CLUSTERS = 3
model = KMeans(n_clusters=NUM_CLUSTERS, max_iter=1000, random_state=True, n_init=50).fit(X=model_w2v.wv.vectors.astype('double'))
```

```python
def extract_words(list_of_tuples):
    list_of_words = []
    for tup in list_of_tuples:
        word = tup[0]
        list_of_words.append(word)
    return list_of_words   
```

```python
for i in range(NUM_CLUSTERS):
    print(extract_words(model_w2v.wv.similar_by_vector(model.cluster_centers_[i], topn=50, restrict_vocab=None)))
    print('\n')
```

```python
def get_synonyms(word):
    words = extract_words(model_w2v.wv.similar_by_vector(word, topn=50, restrict_vocab=None))
    words = ' '.join(words)
    f = open((word+"_synonyms"+".txt"), "a")
    f.write(words)
    f.close()
    print(word)
    print(words + "\n")
```

```python
zeros = np.zeros(100)
for i in range (100):
    zeros[i] = 1
    print(model_w2v.wv.similar_by_vector(zeros, topn=1, restrict_vocab=None)[0])
    zeros[i] = 0
```

```python
get_synonyms("musk")
get_synonyms("destroy")
get_synonyms("tanking")
get_synonyms("elon")
get_synonyms("trump")
get_synonyms("free")
```

```python
from numpy import dot
from numpy.linalg import norm
def cos_sim(a, b):
    return dot(a, b)/(norm(a)*norm(b))
    
```

```python
Sum_of_squared_distances = []
K = range(1,10)
for num_clusters in K :
 kmeans = KMeans(n_clusters=num_clusters, max_iter=1000, random_state=True, n_init=50).fit(X=model_w2v.wv.vectors.astype('double'))
 Sum_of_squared_distances.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Sum of squared distances/Inertia') 
plt.title('Elbow Method For Optimal k')
plt.show()
```
